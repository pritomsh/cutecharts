```python
# import libraries
import pandas as pd
import cutecharts.charts as ctc
```

```python
# Load data
data = pd.read_csv('data/Customer-Churn-Prediction.csv')
data.head()
```

```python
newdata = data.groupby('PaymentMethod').size().rename_axis('PaymentMethodCategory').reset_index(name='Count')
newdata.head()
```

```python
# pie chart 
pie = ctc.Pie('Payment Method Category', # title
              width='720px',height='720px')

# set the chart options
pie.set_options(labels=list(newdata.PaymentMethodCategory), # country names as labels
                inner_radius=0,                  # inner radius set to 0
                colors=['#ee333c','#168ab5'])

# label to be shown on graph
pie.add_series(list(newdata['Count'])) 

# display the charts
pie.render_notebook()
```
