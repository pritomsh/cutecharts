```python
# import libraries
import pandas as pd
import cutecharts.charts as ctc
from cutecharts.faker import Faker
```

```python
# Load data
data = pd.read_csv('data/Customer-Churn-Prediction.csv')
data.head()
```

```python
newdata = data.groupby('PaymentMethod').size().rename_axis('PaymentMethodCategory').reset_index(name='Count')
newdata.head()
```

```python
# pie chart 
pie = ctc.Pie('Payment Method Category', # title
              width='720px',height='720px')

# set the chart options
pie.set_options(labels=list(newdata.PaymentMethodCategory), # country names as labels
                inner_radius=0,                  # inner radius set to 0
                colors=['#ee333c','#168ab5'])

# label to be shown on graph
pie.add_series(list(newdata['Count'])) 

# display the charts
pie.render_notebook()
```

```python
# pie chart 
pie = ctc.Pie('Payment Method Category', # title
              width='720px',height='720px')

# set the chart options
pie.set_options(labels=list(newdata.PaymentMethodCategory), # country names as labels
                inner_radius=0.8,                  # inner radius set to 0
                colors=['#ee333c','#168ab5'])

# label to be shown on graph
pie.add_series(list(newdata['Count'])) 

# display the charts
pie.render_notebook()
```

```python
colors=['#FFF1C9', '#F7B7A3', '#EA5F89', '#9B3192', '#57167E', '#47B39C','#00529B']

chart = ctc.Bar('Payment Method Category', width='600px', height='200px')

chart.set_options(labels=list(newdata.PaymentMethodCategory),

            x_label='Category',

            y_label='Count',   

            colors=Faker.colors
)

chart.add_series('Geners',list(newdata['Count']))

chart.render_notebook()
```

```python
chart = ctc.Scatter("Helps to gain insights like if movies with higher budget have high popularity", width='700px', height='200px')
chart.set_options(x_label='Popularity',
                  y_label='Budget',
                  dot_size=1,
                  colors=['#47B39C']
 )
chart.add_series('Popularity vs Budget', [(z[0], z[1]) for z in list(zip(data['Churn'],data['tenure']))])
chart.render_notebook()
```

```python
chart = ctc.Radar('Cups of coffee consumed per day')
chart.set_options(
    labels=list(data['Churn']), 
    is_show_legend=True,
    legend_pos='upRight' 
    )
chart.add_series('This Week', list(data['Churn'])) 
chart.add_series('Last Week', list(data['Churn']))
chart.render_notebook()
```
